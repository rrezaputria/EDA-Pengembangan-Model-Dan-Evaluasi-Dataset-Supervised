Melakukan perhitungan manual supervised learning menggunakan metode Naive Bayes langkah demi langkah untuk menyelesaikan permasalahan pada animal dataset yang memiliki fitur sebagai berikut:

a. Jumlah kaki (Kontinyu)

b. Ukuran tubuh (Nominal)

d. Sayap (Nominal)

e. Tanduk (Nominal)

f. Gigi taring (Nominal)

g. Lengan pencapit (Nominal)

Simulasikan perhitungan untuk data uji ke-16 (Badak) dan data uji ke-50 (Kaki Seribu), pisahkan data uji tersebut dari data latih lainnya.

Ikuti tahapan dalam menghitung formula metode Naive Bayes sebagai berikut:

1. Menghitung frekuensi jumlah data dan probabilitas pada tiap kelas/label

2. Menghitung frekuensi jumlah data dan probabilitas tiap fitur pada masing-masing kelas

3. Kalikan semua probabilitas data pada semua fitur dan probabilitas kelas

4. Bandingkan hasil probabilitas antara kelas mamalia dan insekta

5. Kelas yang memiliki nilai probabilitas tertinggi menjadi hasil prediksi kelas pada data yang diuji

# Kode Script Penerapan Naive Bayes Pada Dataset Animal

```python
import pandas as pd

##di lakukan penerapan naive bayes classifer
#dengan langkah-langkah, yakni : 
print("'Di Lakukan Proses Perhitungan Penerapan Naive Bayes Classifer Untuk Data Animal'")


#inisiasi naive bayes class untuk perhitungan probabilitas
class NaiveBayesClass:
    
    
    
    #untuk konstruktor kelas 
    def __init__(self, DataAnimal):
        self.Data = pd.read_csv(DataAnimal, sep = ";") 
        self.DataLatih = self.Data.drop(index = [15, 49]) #inisiasi data latih

        self.JumlahMamalia = self.DataLatih[self.DataLatih["Kelas"] == "Mamalia"].shape[0] #jumlah data kelas
        self.JumlahInsekta = self.DataLatih[self.DataLatih["Kelas"] == "Insekta"].shape[0] #jumlah data kelas 
        self.JumlahDataLatih = self.DataLatih.shape[0]

        self.ProbabilitasMamalia = self.JumlahMamalia / self.JumlahDataLatih #perhitungan probabilitas mamalia
        self.ProbabilitasInsekta = self.JumlahInsekta / self.JumlahDataLatih #perhitungan probabilitas insekta
        
        
        
    #untuk perhitungan probabilitas tiap fitur
    def PerhitunganProbabilitasFitur(self, Fitur, NilaiFitur, Kelas):
        #perhitungan muncul fitur dalam kelas
        JumlahFiturKelas = self.DataLatih[(self.DataLatih["Kelas"] == Kelas) & (self.DataLatih[Fitur] == 
                                                                                NilaiFitur)].shape[0] 
        if JumlahFiturKelas > 0: #kondisi apakah fitur ditemukan/tidak
            #jika iya, di lakukan perhitungan probabilitas fitur
            JumlahDataKelas = self.DataLatih[self.DataLatih["Kelas"] == Kelas].shape[0]
            ProbabilitasFiturKelas = JumlahFiturKelas / JumlahDataKelas
        else:
            JumlahDataKelas = self.DataLatih[self.DataLatih["Kelas"] == Kelas].shape[0]
            ProbabilitasFiturKelas = 0.0 #jika tidak, di lakukan inisiasi nilai probabilitas 0,0 
        return ProbabilitasFiturKelas
    
    
    
    #untuk perhitungan probabilitas data uji
    def HitungProbabilitasDataUji(self, DataUji, ProbabilitasKelas, Kelas):
        #inisiasi awal, bahwa probabilitas data adalah probabilitas kelas 
        ProbabilitasData = ProbabilitasKelas 
        for Fitur in DataUji.columns[2:-1]: 
            NilaiFitur = DataUji[Fitur].values[0] #mengambil nilai fitur dari data uji
            #perhitungan probabilitas fitur kelas
            ProbabilitasFiturKelas = self.PerhitunganProbabilitasFitur(Fitur, NilaiFitur, Kelas) 
            ProbabilitasData *= ProbabilitasFiturKelas
            print(f"Probabilitas {Fitur} : {NilaiFitur} Di kelas {Kelas} : {ProbabilitasFiturKelas}")
        return ProbabilitasData
    
    
    
    #untuk perhitungan prediksi kelas 
    def PrediksiKelas(self, DataUji, ProbabilitasMamalia, ProbabilitasInsekta):
        #perhitungan probabilitas kelas mamalia
        ProbabilitasUjiMamalia = self.HitungProbabilitasDataUji(DataUji, ProbabilitasMamalia, "Mamalia") 
        #perhitungan probabilitas kelas insekta
        ProbabilitasUjiInsekta = self.HitungProbabilitasDataUji(DataUji, ProbabilitasInsekta, "Insekta") 

        #penampilan hasil nilai data uji
        print("'Di Peroleh Nilai Dan Perbandingan Untuk Data Uji'")
        print(f"Di Peroleh Nilai Probabilitas Data Uji Terhadap Kelas Mamalia : {ProbabilitasUjiMamalia}")
        print(f"Di Peroleh Nilai Probabilitas Data Uji Terhadap Kelas Insekta : {ProbabilitasUjiInsekta}")

        if ProbabilitasUjiMamalia > ProbabilitasUjiInsekta: #pembuatan kondisi
            return "Mamalia"
        else:
            return "Insekta"
        
        
        
#untuk menjalankan kode script
if __name__ == "__main__":
    DatasetAnimal = NaiveBayesClass("animal_morphology_classify.csv") #inisiasi objek
    DataUjiBadak = DatasetAnimal.Data.iloc[[15]]
    DataUjiKakiSeribu = DatasetAnimal.Data.iloc[[49]]

    #penampilan hasil probabilitas di tiap kelas
    print("\n'Perhitungan Probabilitas Di Masing-Masing Kelas'")
    print(f"Di Peroleh Jumlah Total Kelas Mamalia Sebesar : {DatasetAnimal.JumlahMamalia}")
    print(f"Di Peroleh Jumlah Total Kelas Insekta Sebesar : {DatasetAnimal.JumlahInsekta}")
    print(f"Di Peroleh Jumlah Total Data Latih Sebesar : {DatasetAnimal.JumlahDataLatih}")
    print(f"Di Peroleh Nilai Probabilitas Kelas Mamalia Sebesar : {DatasetAnimal.ProbabilitasMamalia}")
    print(f"Di Peroleh Nilai Probabilitas Kelas Insekta Sebesar : {DatasetAnimal.ProbabilitasInsekta}")

    #penampilan hasil probabilitas fitur di tiap kelas
    print("\n'Perhitungan Probabilitas Fitur Di Masing-Masing Kelas'")
    print("\n'Nilai Probabilitas Masing-Masing Fitur Pada Setiap Kelas Untuk Data Uji Ke-16 Badak'")
    
    #prediksi data uji ke-16 "badak"
    HasilPrediksiBadak = DatasetAnimal.PrediksiKelas(DataUjiBadak, DatasetAnimal.ProbabilitasMamalia, 
                                                     DatasetAnimal.ProbabilitasInsekta)
    print(f"Hasil Prediksi Data Uji Ke-16 'Badak' : {HasilPrediksiBadak}")

    #prediksi data uji ke-50 "kaki seribu"
    print("\n'Nilai Probabilitas Masing-Masing Fitur Pada Setiap Kelas Untuk Data Uji Ke-50 Kaki Seribu'")
    HasilPrediksiKakiSeribu = DatasetAnimal.PrediksiKelas(DataUjiKakiSeribu, DatasetAnimal.ProbabilitasMamalia, 
                                                          DatasetAnimal.ProbabilitasInsekta)
    print(f"Hasil Prediksi Data Uji Ke-50 'Kaki Seribu' : {HasilPrediksiKakiSeribu}")
```
